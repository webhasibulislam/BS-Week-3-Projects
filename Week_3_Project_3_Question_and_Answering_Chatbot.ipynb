```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set the model to evaluation mode
model.eval()
```

```python
def chatbot(prompt, user_input, chat_history=[], max_history=3, max_new_tokens=50):

    # Add the user's input to the history
    chat_history.append(("User", user_input))

    # Prepare the input text for the model
    input_text = prompt + "\n"

    # Add the chat history to the input text
    for i, (role, message) in enumerate(chat_history[-max_history:]):  # Include only the last `max_history` messages
        input_text += f"{role}: {message}\n"

    # Add the AI's role to the input text
    input_text += "AI:"

    # Tokenize the input text
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # Generate a response
    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,  # Control the length of the generated response
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            top_p=0.95,
            top_k=50,
            temperature=0.7,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # Decode the generated response
    response = tokenizer.decode(output[0], skip_special_tokens=True)

    # Extract only the AI's response
    response = response.split("AI:")[-1].strip()

    # Add the AI's response to the history
    chat_history.append(("AI", response))

    # Return the response and updated history
    return response, chat_history
```

```python
# Define the prompt
prompt = "You are a helpful AI assistant. Answer the user's query concisely and provide examples where possible."

# Initialize chat history
chat_history = []

print("Chatbot is ready! Type 'exit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("AI: Goodbye!")
        break

    # Get the chatbot's response and updated history
    response, chat_history = chatbot(prompt, user_input, chat_history, max_new_tokens=100)

    # Print the chatbot's response
    print("AI:", response)
```
